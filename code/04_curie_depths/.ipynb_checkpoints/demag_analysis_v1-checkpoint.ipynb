# 1. Heat Generation from Decay of Radioactive Elements in Crust

## 1.1. Attempt 1: Calculate from GRS Data

```python
def Q_c(t, C):
    '''
    Calculate crustal heat production in the lithosphere from the following radiogenic heat producing elements: 
        [U238, U235, Th232, K40]
    
    params:
        t is time since present day (yrs)
        C is concentrations of radiogenic elements
    
    sources:
        Hahn et al 2011: Martian surface production and crustal heat flow from Mars Odyssey Gamma-Ray spectrometry
        Turcotte and Schubert 2001: Geodynamics (textbook)    
    '''
    
    f = [0.992742, 0.007204, 1.00, 0.00012] # isotopic fraction -- from Hahn et al 2011
    H = [9.46e-5, 5.69e-4, 2.64e-5, 2.92e-5] # heat release constant, i.e. net energy per unit mass [W/kg] -- from Turcotte and Schubert, 2001
    tau = [4.47e9, 7.04e8, 1.40e10, 1.25e9] # half life [yrs] -- from Turcotte and Schubert, 2001
    
    
    Q_c = 0
    for i in range(len(f)):
        Q_c += f[i] * C[i] * H[i] * math.exp( (t * math.log(2)) / tau[i] )
    return Q_c # * 1e-6
```

```python
C = [-1, -1, 0.6, 0.2]
C[0:2] = [C[2] / 3.8]*2

print(Q_c(0, C))

# t = np.arange(3.6e9,0,-1e5)
t = np.linspace(4e9, 0)
Q = []
for ti in t:
    Q.append(Q_c(ti,C))
    
    

fig, ax = plt.subplots(figsize=(7,5))
ax.invert_xaxis()
plt.plot(t,Q)
```

# 2. Heat-Depth Profile

```python
q_B = 7 # basal / crustal component of heat flow [mW m^-2] varies between 1.3 and 13.5, average of 7
k = 2.5 # thermal conductivity of crust [W m^-1 K^-1]
rho = 2800 # density of crust [kg m^-3]

# calculate crustal heat production
method = 1
match method:
    case 1:
        H = Q_c(0, C)
    case 2:
        crater_lon = 23.45
        crater_lat = 10.79
        H = H_cr[int(lat2matfile(crater_lat)), int(lon2matfile(crater_lon))]    


def T(z):
    return 160+(q_B/k)*z+((rho*H)/(2*k))*z**2
```

```python
depth = np.linspace(0, 120)
temp = [T(z) for z in depth]

fig, ax = plt.subplots(figsize=(7,5))
# ax.invert_xaxis()
plt.title("Curie Depths at Henry Crater due to Radiogenic HPEs\n")
ax.set_ylabel('Temperature (K)')
ax.set_xlabel('Depth (km)')
ax.plot(depth,temp)

ax.hlines(y=580, xmin=depth[0], xmax=94, colors='r', linestyles='--', lw=1)
ax.vlines(x=94, ymin=temp[0], ymax=580, colors='r', linestyles='--', lw=1)

ax.hlines(y=320, xmin=depth[0], xmax=44, colors='r', linestyles='--', lw=1)
ax.vlines(x=44, ymin=temp[0], ymax=320, colors='r', linestyles='--', lw=1)


ax.hlines(y=670, xmin=depth[0], xmax=108, colors='r', linestyles='--', lw=1)
ax.vlines(x=108, ymin=temp[0], ymax=670, colors='r', linestyles='--', lw=1)

# ax.hlines(y=580, xmin=depth[0], xmax=depth[-1], colors='r', linestyles='--', lw=1)
# ax.hlines(y=320, xmin=depth[0], xmax=depth[-1], colors='r', linestyles='--', lw=1)
# ax.hlines(y=670, xmin=depth[0], xmax=depth[-1], colors='r', linestyles='--', lw=1)

plt.show()
```
